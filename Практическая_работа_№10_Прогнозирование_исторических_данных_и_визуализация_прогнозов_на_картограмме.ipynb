Количество элементов ряде данных должно превышать 100. (динамика заболеваний по месяцам за несколько десятилетий, курс валюты и тд, чем больше данных для обучения, тем лучше.)

```python
import pandas as pd
import numpy as np
import plotly.express as px


from google.colab import drive
drive.mount('/content/drive')
data = pd.read_csv('/content/drive/My Drive/Colab Notebooks/bigmac.csv')

data.head()
```

```python
px.choropleth(
  data,
  locations="iso_a3",
  color="dollar_price",
  hover_name="name",
  animation_frame="date",
  color_continuous_scale=px.colors.sequential.Plasma,
  projection="natural earth"
             )
```

```python
data['date'] = pd.to_datetime(data['date'])
data = data.set_index('date')
data
```

```python
data['year'] = data.index.year
data['month'] = data.index.month
```

```python
data = data.drop(columns=['USD_raw', 'EUR_raw', 'GBP_raw', 'JPY_raw', 'CNY_raw', 'GDP_dollar', 'adj_price', 'USD_adjusted', 'EUR_adjusted', 'GBP_adjusted', 'JPY_adjusted', 'CNY_adjusted'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from datetime import timedelta

train_size = int(len(data) * 0.8)
train, test = data[0:train_size], data[train_size:]

model = RandomForestRegressor()
```

```python
unique_countries = data['iso_a3'].unique()
```

```python
for country in unique_countries:
    country_df = data[data['iso_a3'] == country]

    train_size = int(len(country_df) * 0.8)
    train, test = country_df.iloc[:train_size], country_df.iloc[train_size:]

    X_train, y_train = train[['year', 'month']], train['dollar_price']
    model.fit(X_train, y_train)

    last_date = country_df['year'].max()
    next_year_date = last_date + 1
    next_year = pd.DataFrame(index=pd.date_range(start=last_date, end=next_year_date, freq='D'))
    next_year['year'] = next_year.index.year
    next_year['month'] = next_year.index.month
    next_year['iso_a3'] = country
    next_year['dollar_price'] = model.predict(next_year[['year', 'month']])

    data = pd.concat([data, next_year], sort=False)
```

```python
px.choropleth(
  data,
  locations="iso_a3",
  color="dollar_price",
  hover_name="iso_a3",
  animation_frame="year",
  color_continuous_scale=px.colors.sequential.Plasma,
  projection="natural earth"
             )
```
